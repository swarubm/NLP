```python
import re
```

## 1.finding pattern with a specific ending 

```python
 re.findall("at" , " cat mat kat dat dfe sas sfe atgr  ")
```

```python
 re.findall(".at" , " cat mat kat dat dfe sas sfe atgr  ") #1 dot
```

```python
 re.findall("..at" , " cat mat kat dat dfe sas sfe atgr  ")  #2 dot
```

```python
re.findall('9','this is a number 9')
```

## 2 start and end pattern matching

```python
re.findall('\d','this is a number 9 5  6') #returns only number 
```

```python
re.findall('\d$','this is a number 9 5  6')  #returns last element if it is number
```

```python
re.findall('\d$','this is a number 9 5  6 is not')#last element is not a number 
```

```python
re.findall('\d$','this is a number 9 5  is 86')#returns only  digits number is in last 
```

```python
re.findall('\d\d$','this is a number 9 5  is 86')#returns only if two digits numbers are in last 
```

```python
re.findall('^\d','5 this is a number 9 5  is 86')#^returns the first element 
```

## 3 Finding capital letters 

```python
re.findall('[a-z]','Welcome to Sardar')
```

## 4 Finding small letters 

```python
re.findall('[A-Z]','Welcome to Sardar')
```

```python
re.findall('[A-Za-z]','Welcome to Sardar')
```

```python
re.findall('[A-Za-z0-9]','Welcome to Sardar 9')
```
