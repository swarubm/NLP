```python
import re
mail = 'swaroopoffice@office.com'
p = '[A-Za-z0-9]+@[\w]+.[\w]+'
re.search(p,mail)
```

```python
re.search(p,'swaroop@office.in').group()
```

```python
re.search(p,'swaroop@office.in  randomlu').group()
```

```python
re.search(p,'swaroop@of random fice.in  randomlu').group()
```

## email id with specific domain

```python
mail = 'swaroop@drttit.org'
p = '[A-Za-z0-9]+@(drttit).(org)$'
re.search(p, mail)
```

```python
re.search(p, 'sardar@drttit.org')
```

```python
re.search(p, 'swaroop.sardar@drttit.org')
```

## Email with multiple domains 

```python
mail = 'swaroop@drttit.org'

p = '[A-Za-z0-9]+@[A-Za-z]+.(org|edu|gov|net|in)'

re.search(p, mail)
```

```python
re.search(p, 'sardar@drttit.in')
```
