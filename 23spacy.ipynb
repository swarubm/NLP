# synonym

```python
synonyms=[]
```

```python
synonyms=[]
for s in wordnet.synsets("Happy"):
    for lemma in s.lemmas():
        synonyms.append(lemma.name())
```

```python
print(synonyms)
```

```python
synonyms=[]
for s in wordnet.synsets("Strong"):
    for lemma in s.lemmas():
        synonyms.append(lemma.name())
print(synonyms)
```

# antonym

```python
ant=[]
for a in wordnet.synsets('Healthy'):
    for lemma in a.lemmas():
        if lemma.antonyms():
            ant.append(lemma.antonyms()[0].name())
print(ant)
```
