## 3.Find iter

```python
for s in re.finditer("are", "are you in my area ,  aren't aint you? "):
    print(s.span())
```

## 4.creating regular expression

```python
txt = 'My telephone number is 833-3324-343'
pattern = '\d\d\d-\d\d\d\d-\d\d\d'
re.search(pattern, txt)
```

```python
txt = 'My telephone number is 833-3324-343'
pattern = '\w\w\w'
re.search(pattern, txt)
```

```python
txt = 'My telephone number is 834-4324-345'
pattern = '\d{3}-\d{4}-\d{3}'
re.search(pattern, txt)
```

```python
re.search(pattern, txt).group()
```

```python
txt = 'My telephone number is 834-4324-345'
pattern = '\d{3}-\d{4}-\d{3}'
re.search(pattern, txt).group()
```
