## Part-of-speech Tags

```python
import spacy
```

```python
nlp=spacy.load('en_core_web_sm')
```

```python
nlp
```

```python
s=u'Eshant has studied Data Science from GFG and is now making various projects'
```

```python
d=nlp(s)
```

```python
d[2]
```

```python
d[2].text
```

## parts of speech for 3rd word

```python
d[2].pos_
```

```python
d[2].tag_
```

```python
d
```

```python
for token in d:
    print(f"{token.text:{15}}{token.pos_:{15}}{token.tag_:{15}}{spacy.explain(token.tag_)}")
```

```python
d=nlp(u"I like to play cricket.")
```

```python
for token in d:
    print(f"{token.text:{15}}{token.pos_:{15}}{token.tag_:{15}}{spacy.explain(token.tag_)}")
```
