```python
import re
```

## 1 Removing special charecters

```python
txts = " Emerging job hubs: Indore, Coimbatore, Jaipur, Bhubaneswar, and Lucknow are seeing job opportunities grow faster than salaries."
```

```python
re.findall('[^!+]',txts)
```

```python
txt = 'Eshant is ha%ppy! But, wh$y is h*e not playing?'
```

```python
''.join(re.findall('[^!$,%*?]+',txt))
```

## 2.Exclusion

```python
txt = "I'm not gay and 24"
print(''.join(re.findall('\D',txt)))
```

## 3 finding printing all the patterns 

```python
txt = "Hello I am studing from Geeks-for-Geeks and it is amazing. Let's see how this work-it-out, wekds-sfd-sedsa"
```

```python
re.findall('[\w]+-[\w]+-[\w]+', txt)
```

## 4 Phone number pattern matching

```python
for no in ["64-6534-342", "543-5345-645","4563-453-445","53-5453-5345","435-234-6324"]:    
    print(str(re.findall('[\d]+-[\d]+-[\d]' , no)[0]).replace('-',''))
```
