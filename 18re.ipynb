## 1 word finding 

```python
txt = "it covers the tense situation around Iran, the Strait of Hormuz, and ongoing high-stakes talks with the U.S., Pakistan’s mediation role, and wider Middle East conflicts."
```

```python
for i in range(len(txt)):
    print(txt[i:i+6],i)   
```

```python
word  = "is"
```

```python
for i in range(len(txt)):
    if(txt[i:i+len(word)],i):
        print(i,i+len(word))
```

```python
txt = 'AI is having the capability to revolutionize all the industries. The Industry is ready for a big change'.lower()

word = 'the'.lower()
```

```python
for i in range(len(txt)):   
    if(txt[i:i+len(word)] == word):
        print(i,i+len(word))
```
